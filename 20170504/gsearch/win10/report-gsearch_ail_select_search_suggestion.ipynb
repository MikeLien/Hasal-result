```python
import glob
import json 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from os import environ
from IPython.display import display, HTML

key_index = ['avg_time',
            'max_time', 
            'med_time',
            'min_time',
            'std_dev']

#
# Read report filename from environment varilable 'file' 
#
filename = environ['file']
with open(filename) as data_file:
    data = json.load(data_file)
    d = pd.DataFrame(data)  

# drop empty 'time_list'
for c in d:
    if (d[c]['time_list'] == []) :
        d.drop(c, axis=1, inplace=True)
    
# Select columns and rows
d.loc[key_index, :]
```

```python
# Retrive 'run_time' from each run
runtime = pd.DataFrame([pd.DataFrame(d[c]['time_list'])['run_time'] for c in d]).T
runtime.columns = d.columns
runtime
```

```python
# Report percentile
runtime.quantile([0.01, 0.95, 0.99], interpolation='nearest')
```

```python
# Report summary
runtime.describe()
```

```python
# Plot input latency histogram
runtime.plot.hist(bins=20, 
                  histtype='bar', 
                  color=['orange'], 
                  subplots=True, layout=(1,1), 
                  figsize=(20,10))
runtime.plot.box(subplots=True,
                 layout=(1,1),
                 sharey=True, sharex=True, figsize=(20, 10))
plt.show()
plt.show()
```
